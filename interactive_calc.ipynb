```python
from ipywidgets import interact, IntSlider, Layout, Checkbox
from debt_calculator import parse_csv

csv_file = "debts/emily_loans.csv"

def interactive_loan_visualizer(monthly_payment, debt_snowball):
    all_loans = parse_csv(csv_file)
    all_loans.calculate_loans(1000, monthly_payment, showplots="all", ramsay=debt_snowball)

slider = IntSlider(value=4000, min=500, max=10000, step=100, description="Monthly Payment ($)", layout=Layout(width='600px'), style={'description_width': '150px'})
checkbox = Checkbox(value=False, description='Use Ramsay Method (debt-snowball)', indent=False)

interact(interactive_loan_visualizer, monthly_payment=slider, debt_snowball=checkbox)
```
